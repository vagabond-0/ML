```python
from sklearn.feature_extraction.text import TfidfVectorizer
def scale_datasets(dataframe,oversampling):
    text_columns = ["title", "text","subject","date"]
    label_column = "True"
    vectorizer = TfidfVectorizer(max_features=5000)
    text_features = vectorizer.fit_transform(dataframe[text_columns[0]] + " " + dataframe[text_columns[1]])
    x = text_features.toarray()
    y = dataframe[label_column].values
    scaler = StandardScaler()
    x = scaler.fit_transform(x)
    if oversampling:
        res = RandomOverSampler()
        x,y = res.fit_resample(x,y)
    data = np.hstack((x,np.reshape(y,(-1,1))))
    return data,x,y
```

```python
train,xtrain,ytrain = scale_datasets(df,oversampling=True)
valid,xvalid,yvalid = scale_datasets(df,oversampling=False)
test,xtest,ytest = scale_datasets(df,oversampling=False)
```
